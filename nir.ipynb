```python
import matplotlib.pyplot as plt
import numpy as np

import cv2
```

```python
img = cv2.imread('./88_3_forgery.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
light_green = (60, 100, 100)
dark_green = (80, 255, 255)
```

```python
from matplotlib.colors import hsv_to_rgb
lo_square = np.full((10, 10, 3), light_green, dtype=np.uint8) / 255.0
do_square = np.full((10, 10, 3), dark_green, dtype=np.uint8) / 255.0
plt.subplot(1, 2, 1)
plt.imshow(hsv_to_rgb(do_square))
plt.subplot(1, 2, 2)
plt.imshow(hsv_to_rgb(lo_square))
plt.show()
```

```python
hsv_nemo = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

mask = cv2.inRange(hsv_nemo, light_green, dark_green)
result = cv2.bitwise_and(img, img, mask=mask)
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```
